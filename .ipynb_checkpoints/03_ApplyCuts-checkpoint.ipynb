# Apply Cuts to select candidates for GP

This notebook will take DES SN data and apply various cuts to produce an SNID name list, which can be passed to PSNID for fitting and assessment.  The final list can be parsed to MAKE_TEMPLATES/GP2-code-vX.ipynb to produce SED time-series templates for SNANA.

```python
import os.path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy.io import ascii
```

```python
BASEDIR = '/Users/masao/dessn/'
```

# Workflow
**(Previous code 02_PrepData.ipynb)**
* Run SNANA
* Flags
  * `flagSNR5` : from S/N>5 light curve points; 1 = single season, 2 = two adjacent seasons, 0 = otherwise [1,2,0]
  * `yearSNR5` = year with largest `nSNR5_YX` [0,5]
  * `N_ml_flag` = number of seasons with more that 5 epochs that pass ML cuts [0,5]
  * `N_ml_flag2` : 1 = single season, 2 = two adjacent seasons, 0 = otherwise [1,2,0]
  * `Year` = year with largest `N_ml_YX` [0,5]
  * `TFLAG` (transient status)

**(This code)**

* Cut on `TFLAG`, `N_ml_flag`, and `N_ml_flag2`

* Run PSNID
* Apply light curve cuts
  * `peakmag_i < 21.5`
  * `firstdetday_i < 100` (detected before November 9 = 100d from Aug 1)
  * sum of `nnondet_{g,r,i,z} >= 4`
* Apply PSNID cuts
  * CCSN: `PBAYES_Ia_OPTZ0 < 0.5 | FITPROB_Ia_OPTZ0 < 0.1`
  * SNIa: ` PBAYES_Ia_OPTZ0 >= 0.5 & FITPROB_Ia_OPTZ0 >= 0.1`
