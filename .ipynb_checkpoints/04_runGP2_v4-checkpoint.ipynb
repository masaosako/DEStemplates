## Functions useful for reading ascii files 

```python
def getData(file_name,col_num):
    data=[]
    for line in open(file_name).readlines():
        line=line.strip()
        cols=line.split()
        if line.startswith('#'): continue
        else: 
            try:
                data.append(float(cols[col_num]))  
            except: continue
                
    return (data)

def GetData(file_name,col_num):
    data=[]
    for line in open(file_name).readlines():
        line=line.strip()
        cols=line.split()
        if line.startswith('#'): continue
        else: data.append(str(cols[col_num]))
    return (data)
```

## Extinction Calculation 

Below we define a function to calcualte the extinction for a given object at a given redshift.
The equations are taken directly from Cardelli, Clayton, & Mathis (1989) extinction law.  
Here av is defined for the V band whicih is at 5495 Angstroms
rv is assumed as A(V)/E(B-V) (e.g., = 3.1 in the LMC)     
wave = wavlength in angstrom

The user input here is the following, 
1. A file containing the SN ID, redshift, MWEBV - this should have the header of SNID, z, MWEBV such that it can be read by ascii.read - file1.
2. The specific SN ID of interest.
3. The wavlengths the SN is being observed in.


```python
def extinct(mwebv, ID, wave):  
    
#    """First need to read the sn_av file"""
#    input_file = ascii.read(file1)
#
#    CID=np.array(input_file['SNID'])
#    mwebv=np.array(input_file['MWEBV'])
#    
#    key = np.where(CID==int(ID))

#    avu = float(mwebv[key][0])*3.1
    avu = mwebv*3.1
    print("MWEBV: ", mwebv)
    print("AV: ", avu)

    x = 10000.0/wave # inverse wavelength in microns   
    xv = 1.82
    y = x - 1.82 

    #Creating empty arrays in which to store the final data 
    aval=[]
    bval=[]
    
    #Now need to loop through each indavidual wavlength value
    for i in range(len(x)):
        
        # For IR data 
        if (x[i] >= 0.3 and x[i] < 1.1):                                                                                                     
            a =  0.574*pow(x[i],1.61)
            b = -0.527*pow(x[i],1.61)
            aval.append(a)
            bval.append(b)
    
        # For Optical/NIR data 
        elif (x[i] >= 1.1 and x[i] < 3.3):                                                                                            
            a = 1. + 0.17699*y[i] - 0.50447*np.power(y[i],2) - 0.02427*np.power(y[i],3) + 0.72085*np.power(y[i],4) + 0.01979*np.power(y[i],5) - 0.77530*np.power(y[i],6) + 0.32999*np.power(y[i],7)
            b = 1.41338*y[i] + 2.28305*np.power(y[i],2) + 1.07233*np.power(y[i],3) - 5.38434*np.power(y[i],4) - 0.62251*np.power(y[i],5) + 5.30260*np.power(y[i],6) - 2.09002*np.power(y[i],7)
            aval.append(a)
            bval.append(b)
    
        # For UV data 
        elif (x[i] >= 3.3 and x[i] < 8.0):                                                                                                     
            if (x[i] >= 5.9):
                fa = -0.04473*np.power(x[i]-5.9,2) - 0.009779*np.power(x[i]-5.9,3)
                fb =  0.21300*np.power(x[i]-5.9,2) + 0.120700*np.power(x[i]-5.9,3)
            else:
                fa = fb = 0.0
    
            a =  1.752 - 0.316*x[i] - 0.104/(np.power(x[i]-4.67,2) + 0.341) + fa
            b = -3.090 + 1.825*x[i] + 1.206/(np.power(x[i]-4.62,2) + 0.263) + fb
        
            aval.append(a)
            bval.append(b)
    
        elif (x[i] >= 8.0 and x[i] <= 10.0):  # For Far-UV data                                                                                                 
            a = -1.073 - 0.628*(x[i]-8.) + 0.137*np.power(x[i]-8.,2) - 0.070*np.power(x[i]-8.,3)
            b = 13.670 + 4.257*(x[i]-8.) - 0.420*np.power(x[i]-8.,2) + 0.374*np.power(x[i]-8.,3)
        
            aval.append(a)
            bval.append(b)
        else:
            a = b = 0.0
        
            aval.append(a)
            bval.append(b)

    aval = np.array(aval)
    bval = np.array(bval)
    
    RV = 3.1
    extinct = avu*(aval + bval/RV)
    
#    plt.plot(x,extinct,'bx')
#    plt.show()
    
    return extinct
```

# Create light  curve data  files and store information 


Below fnction takes the  FITOPT00?.LCPLOT file and creates indavidual light curve files which can be easier to handle.

```python
def create_lcfiles(lcfile, SNID, mjdlo, mjdhi):
    
    ID = np.array(getData(lcfile,0), dtype=int)
    MJD = np.array(getData(lcfile,1), dtype=float)
    PHASE = np.array(getData(lcfile,2), dtype=float)
    FLUX  = np.array(getData(lcfile,3), dtype=float)
    FLUXERR  = np.array(getData(lcfile,4), dtype=float)
    flag = np.array(getData(lcfile,5), dtype=int)
    filt = np.array(GetData(lcfile,6), dtype=str)
    chi = np.array(getData(lcfile,7), dtype=float)
    fit = np.array(GetData(lcfile,8), dtype=int)
    
    #Now find  where  your object of interst is in the file
    
    key1 = np.where(ID==SNID)

    IDn = ID[key1]
    MJDn  = MJD[key1]
    PHASEn = MJD[key1] - peakmjd
    FLUXn = FLUX[key1]
    FLUXERRn  = FLUXERR[key1]
    flagn = flag[key1]
    filtn = filt[key1]
    chin  = chi[key1]
    fitn  = fit[key1]
    
    f1 = str(SNID)+"_lc.dat"
    
    if os.path.exists(f1):
        os.remove(f1)
        
    results=open(f1,'w')
    res_dat=[]
    
    print("ID        MJD         PHASE    FLUX         FLUXERR    FLAG FILT CHI^2 FIT", file=results)
    
    for a in range(len(IDn)):
        if (MJDn[a] > mjdlo) & (MJDn[a] < mjdhi):
            print(IDn[a], MJDn[a], PHASEn[a], FLUXn[a], FLUXERRn[a], flagn[a], filtn[a], chin[a], fitn[a], file=results)
            res_dat.append(results)
        
    results.close()
    
```

The next fuction obtains the light curve information for each object and stores it.

- Here the user needs to input the directory where the previous lc files are stored, the SN ID/s of interes, SNID's, and the SN type of fit
- Fit option key  word is ufit: 1 = SNIA, 2=SN IBC, 3  = SN II. 

```python
def get_lc_info(snid, mjdlo, mjdhi, ufit):

    #Grabs all lc files within your specified directory
    n1 = glob.glob(str(int(snid))+"_lc.dat")[0]

    lc_file = ascii.read(str(n1))
    lcid=np.array(lc_file['ID'])
    MJD = np.array(lc_file['MJD'])
    PHASE=np.array(lc_file['PHASE'])
    FLUX=np.array(lc_file['FLUX'])
    FLUXERR=np.array(lc_file['FLUXERR'])
    flag=np.array(lc_file['FLAG'])
    filt=np.array(lc_file['FILT'])
    fit=np.array(lc_file['FIT'])
    
    Tot_bands=['g','r','i','z']
    Tot_phase = []
    Tot_mjd = []
    Tot_flux = []
    Tot_flux_err = []
    

    for a in range(len(Tot_bands)):
        dat=np.where((filt==Tot_bands[a]) & (flag==1) & (fit==int(ufit)))
        datf=np.where((filt==Tot_bands[a]) & (flag==-1) & (fit==int(ufit)))

        #Combining the good and the bad data- good is flagged as 1,  bad as -1
        ftot_mjd = np.concatenate((MJD[dat],MJD[datf]), axis=0)
        ftot_phase=np.concatenate((PHASE[dat],PHASE[datf]), axis=0)
        f_in=np.argsort(ftot_mjd)
        ftot_flux=np.concatenate((FLUX[dat],FLUX[datf]), axis=0)
        ftot_fluxe=np.concatenate((FLUXERR[dat],FLUXERR[datf]), axis=0)
        
        #getting unique values
        t_f=np.unique(ftot_mjd[f_in],return_index=True)
        
        fband_mjd0 = ftot_mjd[f_in][t_f[1]]
        fband_phase0 = ftot_phase[f_in][t_f[1]]
        fband_flux0 = ftot_flux[f_in][t_f[1]] 
        fband_flux_err0 = ftot_fluxe[f_in][t_f[1]] 

        # add pre-season epoch
        fband_mjd1 = np.insert(fband_mjd0, 0, mjdlo-60.0)
        fband_phase1 = np.insert(fband_phase0, 0, -100.0)
        fband_flux1 = np.insert(fband_flux0, 0, 0.0000)
        fband_flux_err1 = np.insert(fband_flux_err0, 0, 5.0000)

        # add post-season epoch
        #fband_mjd = np.append(fband_mjd1, mjdhi+100.0)
        #fband_phase = np.append(fband_phase1, 300.0)
        fband_mjd = np.append(fband_mjd1, mjdhi+60.0)
        fband_phase = np.append(fband_phase1, 250.0)
        fband_flux = np.append(fband_flux1, 0.0000)
        fband_flux_err = np.append(fband_flux_err1, 5.0000)

        Tot_mjd.append(fband_mjd)
        Tot_phase.append(fband_phase)
        Tot_flux.append(fband_flux)
        Tot_flux_err.append(fband_flux_err)
        
    
    return Tot_bands, Tot_mjd, Tot_phase, Tot_flux, Tot_flux_err
    
```

# The Gaussian Process

The next few functions build the data in the right form for the  gaussian process and then  apply the process itself.

```python
def combine_files(CC_ID, mjdlo, mjdhi):
    """We want to combine the files to make them usable in the 2D GP"""
    
    # These are the cental bandpasses for the DES griz bands."""
    central_wave = [4709.40, 6286.09, 7708.05, 9134.05]
    Tot_bands, Tot_mjd, Tot_phase, Tot_flux, Tot_flux_err = get_lc_info(CC_ID, mjdlo, mjdhi, ufit=2)
    
    #We want to concatinate all the data into one array

    T1m = np.concatenate((Tot_mjd[0], Tot_mjd[1]), axis=0)
    T2m = np.concatenate((T1m, Tot_mjd[2]), axis=0)
    Tot_mjd_all = np.concatenate((T2m, Tot_mjd[3]), axis=0)

    T1f = np.concatenate((Tot_flux[0], Tot_flux[1]), axis=0)
    T2f = np.concatenate((T1f, Tot_flux[2]), axis=0)
    Tot_flux_all = np.concatenate((T2f, Tot_flux[3]), axis=0)

    T1fe = np.concatenate((Tot_flux_err[0], Tot_flux_err[1]), axis=0)
    T2fe = np.concatenate((T1fe, Tot_flux_err[2]), axis=0)
    Tot_fluxe_all = np.concatenate((T2fe, Tot_flux_err[3]), axis=0)

    gb = [4709.40]*len(Tot_mjd[0])
    rb = [6286.09]*len(Tot_mjd[1])
    ib = [7708.05]*len(Tot_mjd[2])
    zb = [9134.05]*len(Tot_mjd[3])

    T1b = np.concatenate((gb, rb), axis=0)
    T2b = np.concatenate((T1b, ib), axis=0)
    Tot_band_all = np.concatenate((T2b, zb), axis=0)
    
    # Want to sort by mjd and get the indices so you can make them all correct
    sort_ind  = np.argsort(Tot_mjd_all)

    Tot_mjd_alls = Tot_mjd_all[sort_ind]
    Tot_flux_alls  = Tot_flux_all[sort_ind]
    Tot_fluxe_alls  = Tot_fluxe_all[sort_ind]
    Tot_band_alls  = Tot_band_all[sort_ind]
    
    return Tot_mjd_alls, Tot_flux_alls, Tot_fluxe_alls, Tot_band_alls
```

```python
def convert2mags(Tot_flux_alls, Tot_fluxe_alls):
    
    #Tot_mag_alls  = 27.5 - 2.5*np.log(Tot_flux_alls)
    #Tot_mage_alls = 1.08573616*(Tot_fluxe_alls/Tot_flux_alls)
    
    Tot_mag_alls  = np.where(Tot_flux_alls <= 1.0, 27.50, 27.5 - 2.5*np.log10(Tot_flux_alls))
    Tot_mage_alls = np.where(Tot_flux_alls <= 1.0,  1.00, 1.08573616*(Tot_fluxe_alls/Tot_flux_alls))
    
#    mags  = 27.5 - 2.5*np.log10(pred_gp2[d][c])  #Rick's AB mag
#    mager = (2.5/2.302585093)*(err_gp2[d][c]/pred_gp2[d][c]) #error  on  mag
    
    return Tot_mag_alls, Tot_mage_alls
```

```python
def convert2flux(Tot_mag_alls, Tot_mage_alls):
    
    #Tot_flux_alls  = np.power(10.0, -0.4*(Tot_mag_alls - 27.5))
    Tot_flux_alls  = 10.0**(-0.4*(Tot_mag_alls - 27.5))
    Tot_fluxe_alls = Tot_mage_alls*Tot_flux_alls/1.08573616

    return Tot_flux_alls, Tot_fluxe_alls
```

```python
def gp_2d_long(guess_length_scale, mjdall, fluxall, flux_errall, bands):

    """This is  all  taken from Avacado -  
    see https://github.com/kboone/avocado/blob/master/avocado/astronomical_object.py
    In this a 2D matern kernal is used  to  model the transient. The kernel
    width in the wavelength direction is fixed. We fit for the kernel width
    in the time direction"""

    #Want to compute the scale factor that we will use...
    signal_to_noises = np.abs(fluxall) / np.sqrt(flux_errall ** 2 + (1e-2 * np.max(fluxall)) ** 2)
    scale = np.abs(fluxall[np.argmax(signal_to_noises)])
#    print("signal_to_noises: ", signal_to_noises)
    print("GP scale: ", scale)
    print("GP guess_length_scale: ", guess_length_scale)
    
    #This is the two dimensional kernal
    kernel = (0.5 * scale) ** 2 * kernels.Matern32Kernel([guess_length_scale ** 2, 6000 ** 2], ndim=2)

    gp = george.GP(kernel)
    guess_parameters = gp.get_parameter_vector()

    x_data = np.vstack([mjdall, bands]).T
    gp.compute(x_data, flux_errall)

    def neg_ln_like(p):
        gp.set_parameter_vector(p)
        return -gp.log_likelihood(fluxall)

    def grad_neg_ln_like(p):
        gp.set_parameter_vector(p)
        return -gp.grad_log_likelihood(fluxall)

    bounds = [(0, np.log(1000 ** 2))]
    bounds = [(guess_parameters[0] - 10, guess_parameters[0] + 10)] + bounds
    #If i include  the bounds it seems  to  mess  everything up.

    fit_result = minimize(neg_ln_like, gp.get_parameter_vector(), jac=grad_neg_ln_like)  
    gp.set_parameter_vector(fit_result.x)
    gaussian_process = partial(gp.predict, fluxall)
    
    return gaussian_process
```

```python
def gauss_dat_long(gp, snid, cw_start, cw_end, cadence, wave_sep, mjdall):

    #This makes a new array of time in bins of cadence
    times = np.arange(min(mjdall),max(mjdall)+1,cadence)  

    central_wave  = np.arange(cw_start, cw_end, wave_sep)

    # print(times)
    # print(central_wave)
    
    predictions_full = []
    prediction_uncertainties_full = []
    
    for a in  range(len(central_wave)):
        wavelengths = np.ones(len(times)) * central_wave[a]
        pred_x_data = np.vstack([times, wavelengths]).T
        band_pred, band_pred_var = gp(pred_x_data, return_var=True)
        prediction_uncertainties = np.sqrt(band_pred_var)
        predictions = np.array(band_pred)
        
        #Note that the flux here has not been corrected for anything - des flux
        predictions_full.append(predictions)
        prediction_uncertainties_full.append(prediction_uncertainties)
      
    
    predictions_full = np.array(predictions_full)
    prediction_uncertainties_full = np.array(prediction_uncertainties_full)
    
    
    return(predictions_full, prediction_uncertainties_full, times, central_wave)
```

```python
#Now want to make up new plots (log version)
def GPepochRange(ID, mjd_gp, pred_gp, err_gp, central_wave, wavestep, peakmjd, redshift):
     
    # restframe epoch
    mjd_gp   = mjd_gp/(1.0+redshift)

    col = ['b','c','m','r']
    filt = ['g','r','i','z']
    
    # These just help you find the points closest to the  griz bands
    ig = np.where(abs(central_wave-4709.40)<wavestep/2.0)[0]
    ir = np.where(abs(central_wave-6286.09)<wavestep/2.0)[0]
    ii = np.where(abs(central_wave-7708.05)<wavestep/2.0)[0]
    iz = np.where(abs(central_wave-9134.05)<wavestep/2.0)[0]

    ind = [int(ig),int(ir),int(ii),int(iz)]
    
    globallo = -300.0
    globalhi = 300.0

    for a in range(len(ind)):
            
        # maximum model flux
        imf = np.argmax(pred_gp[ind[a]])
        mf = np.amax(pred_gp[ind[a]])
        #print("max model flux = %8.4f" % mf, "%d" %imf)
        #print(pred_gp[ind[a],imf])
    
        # use models in epoch range where lower sig-sigma limit of flux is below 5 mags
        sig = 3.0
        fluxlo = pred_gp[ind[a]]/mf - sig*err_gp[ind[a]]/mf
        snr    = pred_gp[ind[a]]/err_gp[ind[a]]
    
        ifound = 0
        for i in range(imf, 0, -1):
            #print("epoch = %8.4f" % mjd_gp[i], 
            #      "relative flux lo = %10.4e" % fluxlo[i])
            if (fluxlo[i] < 1.0e-2) & (snr[i] < 3.0) & (ifound == 0):
                ilo = i
                ifound = 1
            if i == 0:
                ilo = 0
            
        ifound = 0
        for i in range(imf, len(pred_gp[ind[a]])):
            #print("epoch = %8.4f" % mjd_gp[i], 
            #      "relative flux lo = %10.4e" % fluxlo[i])
            if (fluxlo[i] < 1.0e-2) & (snr[i] < 3.0) & (ifound == 0):
                ihi = i
                ifound = 1
            if (ifound == 0) & (i == len(pred_gp[ind[a]])-1):
                ihi = len(pred_gp[ind[a]])-1
            
        print("ilo = %d" % ilo, "epoch lo = %8.4f" % mjd_gp[ilo], 
              "ihi = %d" % ihi, "epoch hi = %8.4f" % mjd_gp[ihi])
        
        if mjd_gp[ilo] > globallo:
            globallo = mjd_gp[ilo]

        if mjd_gp[ihi] < globalhi:
            globalhi = mjd_gp[ihi]
            
    #print("epoch lo = %2d" % globallo, "epoch hi = %2d" % globalhi)
    
    return globallo, globalhi
```

# These next  functions are used to check your SED and convert it back  into  your lightcurve. 


```python
def read_sed(sedfile, redshift):
    
    """This reads in the template SED. This is assumed to be the source spectrum at zero z.
    Input has to be two columns: Wavelength in Ang and monochromatic flux in erg/s/cm2/A.
    Converts wavelength to observed
    Converts flux to observed
    Returns number of grid points"""
    
    z=float(redshift)
    tmp_wave  = np.array(getData(sedfile,1))
    tmp_flux  = np.array(getData(sedfile,2))
    
    wave=[] #observed frame wavelength
    flux=[] #flux per unit frequencey erg/s/cm2/A
    #Could convert to erg/s/cm2/Hz by doing flux*(wave^2/c) 
    #Also check 1+z factor here
    for a in range(len(tmp_wave)):
        wave.append(tmp_wave[a] * (1.0+z))
        flux.append(tmp_flux[a] / (1.0+z)) #This is still in wavelength space
       
    wave_sed = np.array(wave)
    flux_sed = np.array(flux)
    
    nsed=len(tmp_wave)
    
    #output is the wave adjusted sed, flux adjusted sed, and the original Sed 
    return wave_sed, flux_sed, tmp_wave, tmp_flux
```

```python
def get_mag_sed(wave_sed, flux_sed, wave_flt, transmit):
    
    """This function calculates the mag & mjy flux of the sed for a given des filter 
    wave_set is the adjusted sed wavelength in observed
    wave_flt is the obseved filter wavelength
    flux_sed is the sed flux in erg/s/cm^2/Hz
    transmit is the transmission of the band
    
    The need to get linear interpolation of the sed from min wave_flt to max
    Just using a simple 1d interpolation here"""
      
    f1d = interp1d(wave_sed, flux_sed, kind = "cubic")  
    #This calculates the new y values based on the filter
    flux_sed_int = f1d(wave_flt) 
    
    delta_lam = []
    for n in range(len(wave_flt)):
        if n == 0:
            delta_lam.append((((wave_flt[n+2]-wave_flt[n])/2) * wave_flt[n]))
        if (n > 0) and (n < len(wave_flt)-1):
            delta_lam.append((((wave_flt[n+1]-wave_flt[n-1])/2) * wave_flt[n]))
        if n == len(wave_flt)-1:
            delta_lam.append(((wave_flt[n]-wave_flt[n-1]) * wave_flt[n]))
    
    delta_lam = np.array(delta_lam)
    
    flux_sed_mod = []
    for a in range(len(wave_flt)):
        flux_sed_mod.append(flux_sed_int[a] * transmit[a] * delta_lam[a])
    
    flux_sed_mod = np.array(flux_sed_mod) 
    
    #Now need to set up denominator
    R = 3631e-23
    
    denom = []
    for m in range(len(wave_flt)):
        denom.append(R * (3e18/(wave_flt[m]*wave_flt[m])) * transmit[m] * delta_lam[m])
    
    denom = np.array(denom)
    
    area1 = integrate.trapz(flux_sed_mod,wave_flt)
    area2 = integrate.trapz(denom,wave_flt)
    
    mag = -2.5*np.log10(area1/area2)
                  
    mjy = np.power(10.0,(23.9-mag))
    
    return mag, mjy, flux_sed_mod
    
    
```

```python
#Here i go over an example for an indavidual object and how to obtain  it from scratch

#for cc in range(len(SNID)):
for cc in range(0,1):
   
    #1. Assign necessary variables
    print("step 1")
    CID = SNID[cc]
    redshift = zHD[cc]
    mwebv = MWEBV[cc]
    
    thisyear = int(listForGP['Year'][listForGP['CID']==CID])
    mjdlo = mjdrange['lo'][thisyear]
    mjdhi = mjdrange['hi'][thisyear]
    peakmjd = float(listForGP['peakmjd_r'][listForGP['CID']==CID])


    print("i = %2d" % cc, ", CID = %8d" % CID, ", z = %7.4f" % redshift, ", MWEBV = %7.4f" % mwebv,
          ", Year = %1d" % thisyear,
          ", MJDlo = %10.2f" % mjdlo, ", MJDhi = %10.2f" % mjdhi, ", MJDpeak = %10.2f" % peakmjd)
    
    #2. Create the lc file for the object of interest
    print("step 2")
    lcfile= BASEDIR + 'DES_ALL/DESALL_forcePhoto_real_snana_fits/FITOPT000.LCPLOT.SELECTED'
    create_lcfiles(lcfile, CID, mjdlo, mjdhi)

    #3. Obtain the lc information from the file created and store as arrays
    print("step 3")
    Tot_bands, Tot_mjd, Tot_phase, Tot_flux, Tot_flux_err  = get_lc_info(CID, mjdlo, mjdhi, 3)

    #4. Plot the light curve of interest
    print("step 4 SKIP")
    create_lcs(CID, Tot_phase, Tot_flux, Tot_flux_err)

    #5. Combine all the  data together in the appropriate form.
    print("step 5")
    Tot_mjd_alls, Tot_flux_alls, Tot_fluxe_alls, Tot_band_alls = combine_files(CID, mjdlo, mjdhi)


    #####################################
    #5.5 Convert to mags
    Tot_mag_alls, Tot_mage_alls = convert2mags(Tot_flux_alls, Tot_fluxe_alls)
    '''
    for ii in range(len(Tot_mag_alls)):
        print("%12.4f" % Tot_mjd_alls[ii], "%10.4f" % Tot_band_alls[ii],
              "%10.2f" % Tot_flux_alls[ii], "%10.2f" % Tot_fluxe_alls[ii],
              "%8.4f" % Tot_mag_alls[ii], "%8.4f" % Tot_mage_alls[ii])
    '''
    #####################################


    #6. Run the GP
    print("step 6")
    guess_length_scale = 20.0
    gaussian_process = gp_2d_long(guess_length_scale, \
                                  Tot_mjd_alls, Tot_mag_alls, Tot_mage_alls, Tot_band_alls)
    #                              Tot_mjd_alls, Tot_flux_alls, Tot_fluxe_alls, Tot_band_alls)


    #7. Produce GP model in specified wavelength range.
    print("step 7")
    #cw_start  = 1000.0*(1.0+redshift)
    #cw_end = 20020.0*(1.0+redshift)
    cw_start  = 1500.0*(1.0+redshift)
    cw_end = 12020.0*(1.0+redshift)
    timestep = 1.0*(1.0+redshift)
    wavestep = 20.0*(1.0+redshift)
    predictions_full2, prediction_uncertainties_full2, times, central_wave = \
        gauss_dat_long(gaussian_process, CID, cw_start, cw_end, timestep, wavestep, Tot_mjd_alls)
    # predictions_full and prediction_uncertainties_full are matrices of size (wave, time) bins
    print(predictions_full2.shape, prediction_uncertainties_full2.shape)

    #####################################
    #7.5 Reconvert to flux
    predictions_full, prediction_uncertainties_full = convert2flux(predictions_full2, \
                                                                   prediction_uncertainties_full2)
    #####################################
    
    
    #8. Now plot the lc's + the  GP2's and see what phase range you  want your gp2 to cover
    print("step 8 SKIP")
    create_lcs_mod(CID, Tot_mjd, Tot_flux, Tot_flux_err, \
                   times, predictions_full, prediction_uncertainties_full, central_wave, \
                   wavestep)

    
    #9. Reset all negative fluxes to 0.0 and errors to 0.0
    print("step 9")
    pred_gp2 = np.where(predictions_full<0, 0.0, predictions_full) 
    err_gp2  = np.where(prediction_uncertainties_full<0, 0.0, prediction_uncertainties_full)


    #10. Figure out the peak in the i-band restframe -  this will be used to figure out all the phases  
    print("step 10")
    # i-band
    iid = np.where(abs(central_wave-5000.0) < wavestep/2.0)[0][0]
    iband_rest = pred_gp2[int(iid)]/(1+redshift)
    mpeak = np.nanargmax(iband_rest)
    mjd_peak = times[mpeak]
    print("Restframe 5000A peak MJD = %12.4f" % mjd_peak, ", Restframe 5000A peak flux = %8.4f" % mpeak)
    # bolometric
    bolflux = predictions_full.sum(axis=0)
    thisi   = np.nanargmax(bolflux)
    print(times.shape, central_wave.shape)
    print(bolflux.shape, bolflux[thisi], times[thisi], thisi)
    print("Bolometric peak MJD = %12.4f" % times[thisi], ", Bolometric peak flux = %8.4f" % bolflux[thisi])
    #mjd_peak - times[thisi]
    phasebig  = times - mjd_peak
    
    #11. We dont want the SED to be too large or to cover null space.  As such we determine the signal to noise 
    #ratio and figure out which points are above these values.
    print("step 11")
    snr = pred_gp2[int(iid)]/err_gp2[int(iid)]
    snrcut =  np.where(snr>=0.0)[0]
    ccmin = phasebig[snrcut[0]]
    ccmax = phasebig[snrcut[-1]]
    ccmin = -200.0
    ccmax = 200.0
    #Here we are selecting the phase range we want to include in the SED file
    vals = np.where(np.logical_and(phasebig>=float(ccmin), phasebig<=float(ccmax)))[0]
    print("min phase = %8.2f" % float(ccmin), "max phase = %8.2f" % float(ccmax))

        
    #####################################
    #11.5. Plot the GP with the data; deredshift to restframe
    print("step 11.5")
    # linear scale
    create_lcs_mod2(CID, Tot_mjd, Tot_flux, Tot_flux_err, \
                    phasebig, predictions_full, prediction_uncertainties_full, central_wave, \
                    wavestep, mjd_peak, redshift)
    # log scale; normalized to unity at max
    create_lcs_mod3(CID, Tot_mjd, Tot_flux, Tot_flux_err, \
                    phasebig, predictions_full, prediction_uncertainties_full, central_wave, \
                    wavestep, mjd_peak, redshift)
    
    epochlo, epochhi = GPepochRange(CID, phasebig, predictions_full, prediction_uncertainties_full, central_wave, \
                                    wavestep, mjd_peak, redshift)
    print("epoch lo = %8.4f" % epochlo, "epoch hi = %8.4f" % epochhi)
    #####################################
    

    #12. We now want to cacluate magnitudes and covert the flux such that it is in the format suitable for SNANA
    print("step 12")
    flam_new =[]
    flam_err_new =[]
    phasebig_window = []

    mk5 = OUTDIR + str(CID) + '_GP2_SEDrah.dat'

    if os.path.exists(mk5):
        os.remove(mk5)

    results5=open(mk5,'w')
    res_dat5=[]

    for b in range(len(vals)):
        c = vals[b]

        for d in range(len(central_wave)):

            mags  = 27.5 - 2.5*np.log10(pred_gp2[d][c])  #Rick's AB mag

            mager = (2.5/2.302585093)*(err_gp2[d][c]/pred_gp2[d][c]) #error  on  mag

            fnu = np.float_power(10, ((mags + 48.6)/-2.5)) #in  erg/s/cm^2/Hz

            fnu_err = (mager/(2.5/2.302585093))*fnu

            flam = ((fnu/1000)/(central_wave[d]*central_wave[d]))*2.99792458e21
            flam_new.append(flam)

            flam_err =  ((fnu_err/1000)/(central_wave[d]*central_wave[d]))*2.99792458e21
            flam_err_new.append(flam_err_new)

            phasebig_window.append(phasebig[c])

            print(phasebig[c], central_wave[d], flam, flam_err, file=results5)
            res_dat5.append(results5)

    results5.close()

    ph = np.array(getData(OUTDIR + str(CID) + '_GP2_SEDrah.dat',0))
    wa = np.array(getData(OUTDIR + str(CID) + '_GP2_SEDrah.dat',1))
    flam_new = np.array(getData(OUTDIR + str(CID) + '_GP2_SEDrah.dat',2))
    flam_err_new= np.array(getData(OUTDIR + str(CID) + '_GP2_SEDrah.dat',3))


    #13. We now need to remove the extinction
    print("step 13")
    ext = extinct(mwebv, CID, wa)

    flux_Ex =  []
    fluxe_Ex =  []
    for e in range(len(flam_new)):
        flux_Ex.append(flam_new[e]/np.power(10.,-0.4*ext[e]))
        fluxe_Ex.append(flam_err_new[e]/np.power(10.,-0.4*ext[e]))
    flux_Ex = np.array(flux_Ex)
    fluxe_Ex = np.array(fluxe_Ex)


    #14. Now you want to correct for redshift so that it is at zero and apply a DL correction for 10pc
    print("step 14")
    wa_z = []
    flux_Ex_z = []
    fluxe_Ex_z = []
    for f in range(len(wa)):
        wa_z.append(wa[f]/(1.0+redshift))
        flux_Ex_z.append(flux_Ex[f]*(1.0+redshift))
        fluxe_Ex_z.append(fluxe_Ex[f]*(1.0+redshift))

    wa_z = np.array(wa_z)
    flux_Ex_z = np.array(flux_Ex_z)
    fluxe_Ex_z = np.array(fluxe_Ex_z)

    from astropy.cosmology import LambdaCDM
    cosmo = LambdaCDM(H0=70, Om0=0.3, Ode0=0.7)
    DL = cosmo.luminosity_distance(redshift).value #Note that this comes out in Mpc
    DLm  =  (DL*1e6)/10

    #Adjusting phase and flux accordingly....
    Flux_DL = flux_Ex_z * ((1+redshift)*np.power(DLm,2))
    Fluxe_DL = fluxe_Ex_z * ((1+redshift)*np.power(DLm,2))

    
    #15. Print out the SEDcv
    print("step 15")
    output_sed = BASEDIR + 'MAKE_TEMPLATES/DES_' + str(CID) + '.SED'
    output_pdcsv = BASEDIR + 'MAKE_TEMPLATES/DES_' + str(CID) + '.CSV'


    if os.path.exists(output_sed):
        os.remove(output_sed)

    results3=open(output_sed,'w')
    res_dat3=[]


    if os.path.exists(output_pdcsv):
        os.remove(output_pdcsv)

    results4=open(output_pdcsv,'w')
    res_dat4=[]

    
    print("### DES-"+str(CID)+".sed.restframe.dat #########################", file=results3)
    print("#SED created from DES data of SNID"+str(CID), file=results3)
    print("#\n#  SNTYPE: CC\n#  REDSHIFT:", str(redshift)+"\n#  SNTYPE: SN CC\n#  SURVEY: DES", file=results3)
    print("#Converted to rest frame 10 pc away using H0=70 Om=0.3 Ol=0.7", file=results3)
    gg = float(redshift +1)
    print("#flux scaled up by (1+z)*(D_L/10 pc)^2 = ("+str(gg)+")*("+str(DLm)+")^2", file=results3)
    print("#\n#PHASE  LAMBDA_REST  F_LAMBDA F_LAMBDA_ERR", file=results3)

    print("IPHASE,PHASE,LAMBDA_REST,F_LAMBDA,F_LAMBDA_ERR", file=results4)


    for h in range(len(ph)):
        r = ph[h]/(1+redshift)
        m = wa_z[h]
        n = np.nan_to_num(Flux_DL[h]/(1+redshift))
        q = np.nan_to_num(Fluxe_DL[h]/(1+redshift))

        #if (r > -200.0) & (r < 300.0):
        if (r > epochlo) & (r < epochhi):
            print("%8.2f" % r, "%10.2f" % m, "%15.6e" % n, "%15.6e" % q, file=results3)
            res_dat3.append(results3)

            print("%d" % int(r), ",%.2f" % r, ",%.2f" % m, ",%.6e" % n, ",%.6e" % q, file=results4)
            res_dat4.append(results4)
        
    results3.close()
    results4.close()
    print("")
```
